```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
tempo_otimizado = [4006.436, 3469.934, 3807.397, 3546.667, 2991.566, 3036.579]
tempo_nao_otimizado = [3284.657, 3166.449, 3136.728, 3127.309, 3126.416, 3261.720]
labels = ['Tentativa 1', 'Tentativa 2', 'Tentativa 3', 'Tentativa 4', 'Tentativa 5', 'Tentativa 6']

# Inverter a ordem dos dados
tempo_otimizado_inv = tempo_nao_otimizado[::-1]
tempo_nao_otimizado_inv = tempo_otimizado[::-1]
labels_inv = labels[::-1]

# Configuraes da figura e dos eixos
fig, ax = plt.subplots(figsize=(14, 8))

# Calcula as posies para os dados invertidos
bar_positions_optimized = np.arange(len(tempo_otimizado_inv)) * 2  # Espao duplo para agrupamento
bar_positions_non_optimized = bar_positions_optimized + 1  # Barras no otimizadas logo aps as otimizadas
bar_width = 0.8

# Plotagem das barras para tempos otimizados e no otimizados (inverte as cores)
ax.barh(bar_positions_optimized, tempo_otimizado_inv, height=bar_width, color='red', label='No Otimizado')
ax.barh(bar_positions_non_optimized, tempo_nao_otimizado_inv, height=bar_width, color='cyan', label='Otimizado')

# Define os y-ticks para estarem no meio do grupo de barras
middle_of_groups = (bar_positions_optimized + bar_positions_non_optimized) / 2

# Define os rtulos de y-tick para serem os rtulos invertidos
y_ticks_labels_inv = [f'{label}\nOtimizado\n\n{label}\nNo Otimizado' for label in labels_inv]

ax.set_yticks(middle_of_groups)
ax.set_yticklabels(y_ticks_labels_inv)

# Define a cor de fundo e as cores dos rtulos
ax.set_facecolor('#121212')
fig.patch.set_facecolor('#121212')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Adiciona rtulos de dados
for position, value in zip(bar_positions_optimized, tempo_otimizado_inv):
    ax.text(value, position, f'{value:.3f}', va='center', color='white')  # Exibe com trs casas decimais

for position, value in zip(bar_positions_non_optimized, tempo_nao_otimizado_inv):
    ax.text(value, position, f'{value:.3f}', va='center', color='white')  # Exibe com trs casas decimais

# Define o ttulo
ax.set_title('Comparao de Tempos Otimizados e No Otimizados (Invertido)')

# Coloca a legenda fora da rea de plotagem
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)

# Exibe o grfico
plt.show()
```
