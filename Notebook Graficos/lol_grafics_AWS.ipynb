```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
tempo_otimizado = [17263.021, 17321.623, 18284.099, 18204.766, 17781.305, 19296.233]
tempo_nao_otimizado = [17351.009, 17070.094, 17851.005, 17939.349, 18155.636, 18741.832]
labels = ['Tentativa 1', 'Tentativa 2', 'Tentativa 3', 'Tentativa 4', 'Tentativa 5', 'Tentativa 6']

# Inverter a ordem dos dados
tempo_otimizado_inv = tempo_nao_otimizado[::-1]
tempo_nao_otimizado_inv = tempo_otimizado[::-1]
labels_inv = labels[::-1]

# Configurações da figura e dos eixos
fig, ax = plt.subplots(figsize=(14, 8))

# Calcula as posições para os dados invertidos
bar_positions_optimized = np.arange(len(tempo_otimizado_inv)) * 2  # Espaço duplo para agrupamento
bar_positions_non_optimized = bar_positions_optimized + 1  # Barras não otimizadas logo após as otimizadas
bar_width = 0.8

# Plotagem das barras para tempos otimizados e não otimizados (inverte as cores)
ax.barh(bar_positions_optimized, tempo_otimizado_inv, height=bar_width, color='red', label='Otimizado')
ax.barh(bar_positions_non_optimized, tempo_nao_otimizado_inv, height=bar_width, color='cyan', label='Não Otimizado')

# Define os y-ticks para estarem no meio do grupo de barras
middle_of_groups = (bar_positions_optimized + bar_positions_non_optimized) / 2

# Define os rótulos de y-tick para serem os rótulos invertidos
y_ticks_labels_inv = [f'{label}\nOtimizado\n\n{label}\nNão Otimizado' for label in labels_inv]

ax.set_yticks(middle_of_groups)
ax.set_yticklabels(y_ticks_labels_inv)

# Define a cor de fundo e as cores dos rótulos
ax.set_facecolor('#121212')
fig.patch.set_facecolor('#121212')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Adiciona rótulos de dados
for position, value in zip(bar_positions_optimized, tempo_otimizado_inv):
    ax.text(value, position, f'{value:.3f}', va='center', color='white')  # Exibe com três casas decimais

for position, value in zip(bar_positions_non_optimized, tempo_nao_otimizado_inv):
    ax.text(value, position, f'{value:.3f}', va='center', color='white')  # Exibe com três casas decimais

# Define o título
ax.set_title('Comparação de Tempos Otimizados e Não Otimizados (Invertido)')

# Coloca a legenda fora da área de plotagem
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)

# Exibe o gráfico
plt.show()
```
