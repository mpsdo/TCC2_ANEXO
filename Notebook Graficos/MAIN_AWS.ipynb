## View_1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregar o arquivo Excel
file_path = './V1_AWS.xlsx'
excel_data = pd.read_excel(file_path)

# Corrigir o valor anormalmente grande na coluna 'tempo_nao_otimizado'
excel_data.loc[excel_data['tempo_nao_otimizado'] > 1e7, 'tempo_nao_otimizado'] /= 1000

# Filtrar apenas as linhas com tentativas numeradas
filtered_excel_data = excel_data.dropna(subset=['tentativa'])

# Ordenar os dados pela coluna 'tentativa'
filtered_excel_data.sort_values('tentativa', inplace=True)

# Configurar os dados para o grfico
tempo_otimizado = filtered_excel_data['tempo_otimizado'].tolist()
tempo_nao_otimizado = filtered_excel_data['tempo_nao_otimizado'].tolist()
labels = [f'Tentativa {int(t)}' for t in filtered_excel_data['tentativa'].tolist()]

# Configuraes do grfico
x = np.arange(len(labels))  # A quantidade de barras ser baseada no nmero de tentativas
width = 0.35  # Largura das barras

# Criar o grfico
fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width/2, tempo_otimizado, width, label='Tempo Otimizado', color='#4169e1')
rects2 = ax.bar(x + width/2, tempo_nao_otimizado, width, label='Tempo No Otimizado', color='#ff0000')

# Adicionando textos e etiquetas
ax.set_xlabel('Conjuntos de Dados')
ax.set_ylabel('Tempo (em segundos)')
ax.set_title('Comparao de Tempo Otimizado e No Otimizado por Conjunto de Dados')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

# Funo para adicionar uma etiqueta de valor em cada coluna
def autolabel(rects):
    """Adiciona etiquetas de altura em cada barra do grfico."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(int(height)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Chamar a funo autolabel para as duas sries de barras
autolabel(rects1)
autolabel(rects2)

# Ajuste do layout
fig.tight_layout()

# Mostrar o grfico
plt.show()
```
