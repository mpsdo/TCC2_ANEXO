## View_1

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
tempo_otimizado = [17263.021, 17321.623, 18284.099, 18204.766, 17781.305, 19296.233]
tempo_nao_otimizado = [17351.009, 17070.094, 17851.005, 17939.349, 18155.636, 18741.832]
labels = ['Tentativa 1', 'Tentativa 2', 'Tentativa 3', 'Tentativa 4', 'Tentativa 5', 'Tentativa 6']

# Configurações do gráfico
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 8))  # Ajuste no tamanho do gráfico
rects1 = ax.bar(x - width/2, tempo_otimizado, width, label='Tempo Otimizado', color='#4169e1')
rects2 = ax.bar(x + width/2, tempo_nao_otimizado, width, label='Tempo Não Otimizado', color='#ff0000')

# Adicionando textos e etiquetas
ax.set_xlabel('Conjuntos de Dados')
ax.set_ylabel('Tempo (em segundos)')
ax.set_title('Comparação de Tempo Otimizado e Não Otimizado por Conjunto de Dados')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)  # Rotacionando os rótulos
ax.legend()

# Função para adicionar uma etiqueta de valor em cada coluna
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.1f}'.format(height),  # Formatação dos números
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 10),  # Deslocamento vertical aumentado
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)  # Tamanho de fonte reduzido

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

# Mostrando o gráfico
plt.show()
```
